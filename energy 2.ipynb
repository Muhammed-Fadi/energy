```python
x=df1.drop(columns=['cooling load'])      #removing the cool load columns from df1 and store the rest in x
```

```python
x.columns #view columns in x
```

```python
y=df1['cooling load']      #store the cooling load columns from df1 in y (target)
```

```python
y    #calling y
```

```python
from sklearn.feature_selection import SelectKBest,f_regression          #import sklearn
```

```python
sk=SelectKBest(score_func=f_regression,k='all')        #select all features using the f_regrssion scoring function
```

```python
x_new=sk.fit_transform(x,y)       #select the best features from x based on y 
```

```python
selected_columns=x.columns[sk.get_support()]      # get the name of the selected features 
```

```python
selected_columns
```

```python
xselected=x[selected_columns]
```

```python
featurescore=pd.DataFrame({'features':x.columns,'score':sk.scores_}).sort_values(by='score',ascending=False)
```

```python
featurescore
```

scaling
--------------------

```python
from sklearn.preprocessing import StandardScaler      # importing the standardscaler
```

```python
ss=StandardScaler()     # setting up the scaler 
```

```python
xscaled=ss.fit_transform(xselected)      #scaling the selection features 
```

```python
xscaled      #viewing the scaled ones
```

```python
from sklearn.model_selection import train_test_split     #importing train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(xscaled,y,test_size=0.2,random_state=42)         #split the data into training and testing sets
```

```python
print('x_train shape:',x_train.shape)
print('x_test shape:',x_test.shape)
print('y_train shape:',y_train.shape)
print('y_test shape:',y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression   #importing the alogritham---linear regression
```

```python
LR=LinearRegression()    #create a linearregression object
```

```python
LR.fit(x_train,y_train)      #training data
```

```python
ypred_LR=LR.predict(x_test) # predicting
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score      #import mean_absolute_error,mean_squared_error,r2_error
```

```python
mae=mean_absolute_error(y_test,ypred_LR)   #evaluvating predicted values ( mean_absolute_error,mean_squared_error,r2_score)
```

```python
mae
```

```python
mse=mean_squared_error(y_test,ypred_LR)
```

```python
mse
```

```python
r2=r2_score(y_test,ypred_LR)
```

```python
r2
```

```python
from sklearn.svm import SVR 
```

```python
svr=SVR()
```

```python
svr.fit(x_train,y_train)
```

```python
ypred_svr=svr.predict(x_test)     #predicting
```

```python
mae=mean_absolute_error(y_test,ypred_svr)        #evaluvating predicted values ( mean_absolute_error,mean_squared_error,r2_score)
mse=mean_squared_error(y_test,ypred_svr)
r2=r2_score(y_test,ypred_svr)
```

```python
print (mae)      #displaying the predicted value 
print (mse)
print (r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dr=DecisionTreeRegressor()    # create a decisiontree regression object
```

```python
ypred_dr=dr.predict(x_test)    #predicting
```
